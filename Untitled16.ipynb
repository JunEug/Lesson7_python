```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dz_files/dz4/tmdb_5000_movies.csv')
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['budget'], bins=10)
plt.xlabel('Бюджет')
plt.ylabel('Количество фильмов')
plt.title('Распределение бюджета фильмов')
plt.show() 
# Гистограмма позволит наглядно представить, какая доля фильмов имеет определенные бюджеты, а также выявить наиболее типичный диапазон бюджетов.
```

```python
sns.scatterplot(x='revenue', y='popularity', data=df)
plt.xlabel('Доход')
plt.ylabel('Популярность')
plt.title('Связь между популярностью и доходами фильмов')
plt.show()
# Диаграмма рассеяния позволяет исследовать взаимосвязь между двумя непрерывными переменными и выявить возможные закономерности.
```

```python
correlation_matrix = df[['vote_average', 'revenue']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица оценок и доходов фильмов')
plt.show()
# Корреляционная матрица позволяет исследовать взаимосвязь между различными переменными и определить, есть ли взаимосвязь между оценками пользователей и доходами фильмов.
```

```python
sns.boxplot(x='genres', y='runtime', data=df)
plt.xlabel('Жанры')
plt.ylabel('Длительность')
plt.title('Длительность фильмов по жанрам')
plt.xticks(rotation=90)
plt.show()
# Ящик с усами позволяет наглядно представить статистику о минимальной, максимальной, медиане и выбросах в распределении переменной runtime для каждого жанра фильма.
# Я не знаю как оптимизировать, чтоб сделать красиво :c
```
